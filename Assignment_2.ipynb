Q3. Write a Python function to sum 4 inputs from the user.

```python
def sum(size):
    x=[]
    sum = 0
    print("Input",size,"numbers")
    for i in range(size):
        x.append(int(input()))
    x = np.array(x)
    for j in x:
        sum += j
    return sum
        
```

Q5.Write Python Function to get even number from the list. List len must be equal to 8.

```python
def even():
    x = []
    even_numbers = []
    for i in range(1,8):
        x.append(int(input()))
    for j in x:
        if j % 2 == 0:
            even_numbers.append(j)
    return even_numbers
even()
```
